```python

#creating loop to load all bars (50 at a time limit) in radius prescribed

# initial foursquare places API link
url = 'https://api.foursquare.com/v3/places/search?query=bar&ll=39.96505871218218%2C-75.16367498832258&radius=15000&fields=name%2Cgeocodes%2Clocation%2Ccategories%2Crating%2Cprice%2Cmenu&limit=50'

# initialize empty data list
bars = []

# create index to remove unneccessary characters from link later
length_html_end = len('>; rel="next"') # the link comes encased in an html tag <>; rel="next"

while url is not None:
    resp = requests.get(url, headers=options["headers"])
    result = resp.json()
    bars += result['results']
    if 'Link' in resp.headers:
        link = resp.headers['Link']
        url = link[1:-length_html_end] # remove '<' by starting at 1 and '>; rel="next"' by ending at -cut_tail
    else: url = None

print(len(bars)) #should be over 50
```

```python
# return list object back to dictionary for make_nice function, next
dict_str = "{'results': " + str(bars) + ", 'context': {'geo_bounds': {'circle': {'center': {'latitude': 39.96505871218218, 'longitude': -75.16367498832258}, 'radius': 15000}}}}"
dict_object = eval(dict_str)
```

```python
#function to clean up the resulting data - could edit so more fields don't cause an error

def make_nice(data):
    array = []
    
    for element in data.get('results', []):
        nice_element = {
            #"ID": element['categories']['id'],
            "name": element['name'],
            #"price": element['price'],
            #"rating": element['rating'],
            #"address": element['location']['address'],
            "census_block": element['location']['census_block'],
            "zipcode": element['location']['postcode'],
            "latitude": element['geocodes']['main']['latitude'],
            "longitude": element['geocodes']['main']['longitude'],
        }
        array.append(nice_element)
    
    return array
```

```python
#data all cleaned up

nice_data = make_nice(dict_object)
bars_df = pd.DataFrame.from_dict(nice_data)
bars_df.head()
```

```python
# data turned into a geopandas array

geometry = [Point(lon, lat) for lon, lat in zip(bars_df['longitude'], bars_df['latitude'])]
bars_gdf = gpd.GeoDataFrame(bars_df, geometry=geometry, crs="EPSG:4326")
bars_gdf.head()
```

```python
bars_with_hood = bars_with_hood[['name','WEEKDAY','ZillowName','TIME','geometry']]
bars_with_hood = bars_with_hood.rename(columns={'name':'Bar Name','WEEKDAY':'Weekday','TIME':'Time','ZillowName':'Neighborhood Name','geometry':'geometry'})
bars_with_hood['Weekday'] = bars_with_hood['Weekday'].str.title()
```

```python
# Plot the neighborhood boundary
m = neighborhoods.explore(
    style_kwds={"weight": 4, "color": "black", "fillColor": "none"},
    name="Neighborhood boundary",
    tooltip=False
)


# Add the individual tickets as circle markers and style them
bars_with_hood.explore(
    m=m,  # Add to the existing map!
    marker_kwds={"radius": 7, "fill": True, "color": "crimson"},
    marker_type="circle_marker",  # or 'marker' or 'circle'
    name="Tickets",
)

m  # show map
```

```python
#Chloropleth of count per neighborhood
count_by_nhood = bars_with_hood.groupby('Neighborhood Name')['Neighborhood Name'].agg({'count'})
count_by_nhood = count_by_nhood.rename(columns={'count':'Number of Quizzos'})
count_by_nhood.head()
```

```python
count_by_nhood = neighborhoods.merge(count_by_nhood, left_on='ZillowName', right_on='Neighborhood Name')
count_by_nhood.head()
```

```python
count_by_nhood['area'] = count_by_nhood['geometry'].area
count_by_nhood['density'] = count_by_nhood['Number of Quizzos']/count_by_nhood['area']
count_by_nhood.head()
```

```python
neighborhood_names = list(bars_with_hood['Neighborhood Name'].unique())

neighborhood_names[:10]
```

```python
# Create the widget
neighborhoodSelect = pn.widgets.Select(
    value="University City", options=neighborhood_names, name="Neighborhood"
)

neighborhoodSelect
```

```python
def filter_by_neighborhood(data, neighborhood_name):
    """
    Filter data by neighborhood.
    """

    sel = bars_with_hood["Neighborhood Name"] == neighborhood_name
    return data.loc[sel]


def plot_ticket_data(data, neighborhood_name):
    """
    Plot the ticket data and neighborhood boundary on a Folium map.
    """

    # Get the neighborhood geometry
    sel = neighborhoods["ZillowName"] == neighborhood_name
    hood_geo = neighborhoods.loc[sel]

    # Plot the neighborhood boundary
    m = hood_geo.explore(
        style_kwds={"weight": 2, "color": "black", "fillColor": "none"},
        name="Neighborhood boundary",
        tiles="cartodb positron",
        tooltip=False
    )

    # Step 4: Overlay the individual tickets as circle markers and style them
    data.explore(
        m=m,  # Add to the existing map!
        marker_kwds={"radius": 7, "fill": True, "color": "crimson"},
        marker_type="circle_marker",  # or 'marker' or 'circle'
        name="Tickets"
    )

    return m
```

```python
def create_dashboard_1(neighborhood_name):
    """
    Plot trash-related requests for the input neighborhood.
    """
    # Step 1: Filter the tickets data for input neighborhood
    tickets = filter_by_neighborhood(bars_with_hood, neighborhood_name)

    # Step 2: Make the Folium map
    m = plot_ticket_data(tickets, neighborhood_name)

    # Step 3: Return a Folium pane
    # IMPORTANT: set the height!
    return pn.pane.plot.Folium(m, height=600)
```

```python
# Create the dashboard
ticket_dashboard_1 = pn.Column(
    # Top: the title and dropdown widget
    pn.Column("## Quizzo Locations by Neighborhood", neighborhoodSelect),
    # Add a height spacer
    pn.Spacer(height=25),
    # Bottom: the main chart, bind widgets to the function
    pn.bind(create_dashboard_1, neighborhood_name=neighborhoodSelect),
)


ticket_dashboard_1
```
